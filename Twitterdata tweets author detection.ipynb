```
import zipfile
import os #pip install os
import glob
import csv #pip install csv
import pandas as pd
import ipython_genutils #pip install ipython_genutils
from datetime import date, timedelta

#######################################Global Variables########################################

list_files, data, result = [],[],[]
```

```
list_files = [f for f in os.listdir('./') if f.endswith('.xls')]
print(list_files)
```

```
###############For iterating all the files and appending those to Dataframe#######################

for file in list_files:
        df1 = pd.read_excel(file) 
        print("Accessing and Reading\n -",file)
        df1.columns = df1.iloc[0]
        data = list(df1.iloc[0])
        df1 = df1.drop(df1.index[0])
        #Somefiles had different columnames, so appending column names to list and making it index
        df1.set_index(data[0],inplace=True)   
        #Transposing to change the dates provided to columns
        #df1 = df1.transpose()
        #Used to change the index from"Total posts to Country name" which is done by taking the file name
        # name = file.split("-")[0]
        # df1.rename(index={'Total Posts': name},inplace=True)
        result.append(df1)
result = pd.concat(result)
result.to_excel("output.xlsx")
```
