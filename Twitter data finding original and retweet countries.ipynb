```
import pandas as pd
```

```
df1 = pd.read_csv('TweetandReweetInformation_1.csv')
df2 = pd.read_csv('TweetandReweetInformation_2.csv')
df = pd.concat([df1, df2], ignore_index=True)
```

```
df
```

```
df_result = df.groupby(['Tweet_Id','Tweet_Country','Retweet_Country']).count()[['Retweet_Id']]
```

```
df_result.rename(columns={"Retweet_Id":"Number_of_retweets"}, inplace=True)
```

```
df_result
```

```
df_result.to_csv('RetweetCountCountrywise.csv')
```
