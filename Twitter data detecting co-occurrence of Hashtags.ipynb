```
import pandas as pd
import itertools
```

```
df = pd.read_pickle('AllCountryData.pkl')
print(df.shape)
```

```
contents = []
contents = list(df[df['Post Type']!= 'Retweet']['Contents'])
contents = list(set(contents))
print('Length of contents',len(contents))
```

```
hash_tags = []
hash = '#'
for content in contents:
  temp=content.split()
  for i in temp:
    if hash in i:
      hash_tags.append(i)

hash_tags = list(set(hash_tags))
print('Length of hash tags', len(hash_tags))

```

```
combinations = []
combinations = list(itertools.combinations(hash_tags,2))
print('Length of combinations', len(combinations))
```

```
result_df_rows =[]

for combo in combinations:
  count = 0
  temp = list(combo)
  for content in contents:
    if str(temp[0]) in content:
      if str(temp[1]) in content:
        count = count+1
  result_df_rows.append(tuple(str(temp[0]),str(temp[1]), count))
  print('Processing.....')

```

```
result_df_columns=[]
result_a_df_columns = ['Hashtag1', 'Hashtag2','Co-occurrence']
```

```
result_df = pd.DataFrame(result_df_rows, columns=result_df_columns)
result_df.to_pickle('ResultData.pkl')
result_df.to_csv('ResultData.csv')
print('Result data exported')
```
