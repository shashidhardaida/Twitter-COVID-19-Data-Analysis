```python
import pandas as pd
from langdetect import detect
import glob
import os
```

```python
entries = os.listdir('/Users/xinchen/Documents/US/data')
filenames = sorted(entries)
filenames
```

```python
for filename in filenames[150:]:
    data = pd.read_excel('/Users/xinchen/Documents/US/data/'+filename,skiprows=[0])
    languages=[]
    for index in range(len(data)):
        tweet=data['Contents'][index]
        try:
            lang=detect(tweet)
            languages.append(lang)
        except Exception:
            languages.append("none")
    data['language']=languages
    data.to_excel(filename)
    print(filename)
```
